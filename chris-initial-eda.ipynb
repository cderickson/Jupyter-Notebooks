```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_injury = pd.read_csv('InjuryRecord.csv')
df_player = pd.read_csv('PlayerTrackData.csv')
df_plays = pd.read_csv('PlayList.csv')
```

```python
df_playerinfo = df_plays.groupby(['GameID']).agg({'RosterPosition':'last',
                                                  'PlayerDay':'last',
                                                  'PlayerGame':'last',
                                                  'StadiumType':'last',
                                                  'FieldType':'last',
                                                  'Temperature':'last',
                                                  'Weather':'last',
                                                  'PlayType':'last',
                                                  'PlayerGamePlay':'last',
                                                  'Position':'last',
                                                  'PositionGroup':'last'}).reset_index()
df_playerinfo
```

```python
df_merged = pd.merge(df_injury,df_playerinfo,on='GameID',how='left')
df_merged
```

```python
df_merged['PositionGroup'].value_counts().plot(kind='bar')
plt.xlabel('PositionGroup')
plt.ylabel('Frequency')
plt.title('Distribution of PositionGroup')
plt.show()
```

```python
df_merged['FieldType'].value_counts().plot(kind='bar')
plt.xlabel('FieldType (Plays w/ Injuries)')
plt.ylabel('Frequency')
plt.title('Distribution of FieldType (Plays w/ Injuries)')
plt.show()
```

```python
df_plays['FieldType'].value_counts().plot(kind='bar')
plt.xlabel('FieldType (All Plays)')
plt.ylabel('Frequency')
plt.title('Distribution of FieldType (All Plays)')
plt.show()
```

```python
df_merged['BodyPart'].value_counts().plot(kind='bar')
plt.xlabel('BodyPart')
plt.ylabel('Frequency')
plt.title('Distribution of BodyPart')
plt.show()
```

```python
df_merged['Weather'].value_counts()
```

```python
plt.hist([df_merged.PlayerGame], bins=15, label=['PlayerGame'])
plt.xlabel('Game#')
plt.ylabel('Frequency')
plt.title('Histogram of PlayerGame')
plt.show()
```

```python
df_merged[df_merged['BodyPart'] == 'Knee']['FieldType'].value_counts().plot(kind='bar')
plt.xlabel('FieldType (Knee Injuries)')
plt.ylabel('Frequency')
plt.title('Distribution of FieldType (Knee Injuries)')
plt.show()
```

```python
df_merged[df_merged['BodyPart'] == 'Ankle']['FieldType'].value_counts().plot(kind='bar')
plt.xlabel('FieldType (Ankle Injuries)')
plt.ylabel('Frequency')
plt.title('Distribution of FieldType (Ankle Injuries)')
plt.show()
```
