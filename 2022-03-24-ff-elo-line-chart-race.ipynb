## Import libraries and dataset.

```python
import pandas as pd

elo_over_time = pd.read_csv("C:\\Users\\chris\\Documents\\Datasets\\FFL\\elo_over_time.csv")
```

## Format data for use with Flourish. Replace zeroes with empty strings.

```python
elo_over_time = elo_over_time.replace(0, "")
elo_over_time
```

## Transpose dataframe.

```python
elo_over_time = elo_over_time.T
```

## Set "Date" row as the column header row.

```python
elo_over_time.columns = elo_over_time.iloc[0]
elo_over_time = elo_over_time[1:]
elo_over_time
```

## Export dataframe to CSV.

```python
elo_over_time.to_csv("C:\\Users\\chris\\Desktop\\line_race_data.csv", index=True)
```
